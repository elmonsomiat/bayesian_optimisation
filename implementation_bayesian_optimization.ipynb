```python
from one_dim_bayesian_optimisation import BayesianOptimizer
import matplotlib.pyplot as plt
import numpy as np
```

```python
class BayesianOptimizerLoss(BayesianOptimizer):
    def __init__(self):
        super(BayesianOptimizerLoss, self).__init__()

    def calculate_loss_func(self, x, A=2, B=0.32, C=0.5):
        y = x*np.sin(x) + x**2*np.cos(x)
        random_shift =  np.random.uniform(low=0.9, high=1.1)
        return y*random_shift
```

```python
b_opt = BayesianOptimizerLoss()
```

```python
random_numbers = np.linspace(1,8, 1000).T
plt.scatter(random_numbers, b_opt.calculate_loss_func(random_numbers))
```

```python
x = random_numbers.reshape( len(random_numbers), 1)
X = np.array([1, 3.5, 8])
X = X.reshape(len(X), 1)
y =  b_opt.calculate_loss_func(X)

b_opt.run_optimizer(X, y, x, n_steps=15)
```
